```python
import stanza
from polo2 import PoloConfig, PoloDb, PoloCorpus
import os
import pandas as pd
import tqdm
```

```python
# stanza.download('en') # download English model
```

```python
nlp = stanza.Pipeline('en', processors='tokenize,pos')
```

```python
polo_pub = os.environ['POLO_PUB']
```

```python
projects = sorted([dir for dir in os.listdir(polo_pub) if '.' not in dir])
```

```python
def get_corpus(proj):
    config_file = f"{polo_pub}/{proj}/config.ini"
    cfg = PoloConfig(config_file, slug=proj)
    src_file_sep = cfg.ini['DEFAULT']['src_file_sep']
    corpus_file = f"{polo_pub}/{proj}/corpus/corpus.csv"
    if os.path.exists(corpus_file):
        return open(corpus_file, 'r').readlines()
        # corpus_lines = [line.strip().split(src_file_sep) for line in open(corpus_file, 'r').readlines()]
        # return corpus_lines
    else:
        return False
```

```python
corpus_list = get_corpus('anphoblacht')
```

```python
docs = pd.DataFrame([line.strip().split('|') for line in corpus_list[1:]])[3]
```

```python
doc_data = []
for doc in tqdm.tqdm(docs.values):
    doc_data.append(nlp(doc))
```
